### **Строки: методы**

* find() возвращает индекс первого вхождения подстроки или -1, если подстрока не найдена.
* index(), работает аналогично методу find(), но выбрасывает исключение, если подстрока не найдена.

````python
text = "Hello, world!"
index = text.find("world")  # Поиск подстроки "world"
index
````

````python
text.find("wor")
````

````python
text.find("w"), text.find("o")
````

````python
text[text.find('w'):]
````

````python
text.index("j")
````

## **startswith(), endswith() —**

методы, которые позволяют проверить определенное условие относительно строки.

````python
text = "Hello, world!"
starts_with_hello = text.startswith("Hello w")  # Проверка, начинается ли строка с "Hello"
starts_with_hello
````

````python
text = "Hello, world!"
ends_with_world = text.endswith("world!")  # Проверка, заканчивается ли строка с "world"
ends_with_world
````

* Метод ljust() выравнивает строку по левому краю, добавляя пробелы в конец строки, чтобы достичь заданной ширины.
* Метод rjust() выравнивает строку по правому краю, добавляя пробелы в начало строки.
* Метод center() выравнивает строку по центру, добавляя пробелы с обеих сторон.

````python
text = "Hello"
left_aligned = text.ljust(10)  # Выравнивание по левому краю
right_aligned = text.rjust(10)  # Выравнивание по правому краю
centered = text.center(10)
left_aligned, right_aligned, centered
````

## **Удаление символов**

* Метод strip() удаляет пробелы и другие символы из начала и конца строки.
* Метод lstrip() удаляет символы только из начала строки.
* Метод rstrip() удаляет символы только из конца строки.

````python
text = "  Hello, world!  "
stripped_text = text.strip()  # Удаление пробелов из начала и конца строки
left_stripped_text = text.lstrip()  # Удаление пробелов из начала и конца строки
right_stripped_text = text.rstrip()  # Удаление пробелов из начала и конца строки
stripped_text, left_stripped_text, right_stripped_text
````

## **Практическое задание 2**

* Напишите программу, которая запрашивает у пользователя строку и подстроку, а затем находит все вхождения подстроки в строке и выводит их позиции.
* Используйте метод find() для поиска подстроки.
* Если подстрока не найдена, выведите соответствующее сообщение на экран.

* Пример вывода:
* Введите строку: Hello, how are you? How's the weather today?
* Введите подстроку: How
* Подстрока найдена на позициях: 7, 20

````python
строка = input('Введите строку: ') # Hello, How are you? How's the weather today?
````

````python
подстрока = input('Введите подстроку: ') # How
````

````python
строка.find(подстрока)
````

````python
строка.find(подстрока, 8) # второй аргумент говорит о стартовой позиции поиска подстроки
````

````python
индекс_начала_поиска_подстроки = 0
длина_строки = len(строка)
длина_подстроки = len(подстрока)

while индекс_начала_поиска_подстроки < длина_строки:
  новый_индекс = строка.find(подстрока, индекс_начала_поиска_подстроки)
  if новый_индекс == -1: # когда подстрока более не встречается в строке
    break
  print(новый_индекс)
  индекс_начала_поиска_подстроки = новый_индекс + 1
````
