## **Базовое форматирование**

````python
# аргументы по умолчанию
print("Hello {}, your balance is {}.".format("Adam", 230.2346))

# # позиционные аргументы
print("Hello {0}, your balance is {1}.".format("Adam", 230.2346))

# # аргументы ключевые слова
print("Hello {name}, your balance is {блс}.".format(блс=230.2346, name="Adam"))

# # смешанные аргументы
print("Hello {0}, your balance is {blc}.".format("Adam", blc=230.2346))
````

````python
name = input()
balance = input()
print('Hello ', name, ', your balance is ', balance, sep='')
````

## **f-строки**
f-строки позволяют встраивать выражения и переменные прямо в строку.


````python
name = "John"
age = 30
text = f"My name is {name} and I'm {age} years old"
text
````

````python
price = 9.99
quantity = 5
total = price * quantity
text = f"Total: ${total:.2f}"  # Указание точности числа до двух знаков после запятой
text
````

````python
name = "John"
age = 305
text = f"{name:?>10} is {age:=^10} years old"
# имя прибитое к правому краю в рамках 10 букв, заполненных пробелами
# возраст в рамках 10 букв, заполненных знаком равно
text
````
