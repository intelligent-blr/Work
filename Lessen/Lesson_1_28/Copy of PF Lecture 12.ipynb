## **Цикл for —** это цикл, который используется для выполнения повторяющихся операций над элементами последовательности. Позволяет пройти по всем элементам списка, строки, кортежа или другой итерируемой структуры данных.

````python
for i in range(5):
    print(i)
````

## **range() —** это функция, которая возвращает последовательность чисел от начального значения до конечного значения с определенным шагом. Пример: range(1, 10, 2) вернет последовательность [1, 3, 5, 7, 9].

````python
list(range(103, 106))
````

````python
list(range(10, 0, -1))
````

````python
list(range(0, 10 , 2))
````

````python
# empty range
print(list(range(0)))

# using range(stop)
print(list(range(10)))

# using range(start, stop)
print(list(range(1, 10)))
````

## **break, continue, else в цикле for**

* В цикле for можно использовать операторы break и continue для управления ходом выполнения цикла.
* Оператор break прерывает выполнение цикла и переходит к следующей инструкции после цикла.
* Оператор continue пропускает текущую итерацию цикла и переходит к следующей.
* Цикл for может иметь блок else, который выполняется после полного прохода по итерируемому объекту, т.е. если цикл не был прерван оператором break.




## **enumerate —** это функция, которая позволяет получить индекс и значение элемента последовательности на каждой итерации цикла. Возвращает объект-итератор, содержащий пары (индекс, значение).


````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

````python
list(enumerate(fruits))
````

````python
fruits = ["apple", "banana", "cherry"]
animals = ['pig', 'monkey', 'hamster', 'cat']
````

````python
list(enumerate(fruits))
````

````python
for index, fruit in enumerate(fruits):
    print(animals[index], 'loves', fruit)
````

````python
list(zip(animals, fruits))
````

````python
for element in zip(animals, fruits):
    print(element[0], 'loves', element[1])
````

zip() — это функция, которая позволяет объединить элементы нескольких последовательностей в кортежи. Возвращает итератор, который можно использовать в цикле for. Если последовательности имеют разную длину, zip() остановится на самой короткой.

````python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(name, age)
````

## **with .. as —** это  контекстный менеджер, который автоматически открывает и закрывает файл, обеспечивая правильное управление ресурсами. В этом случае нет необходимости явно вызывать метод close().

````python
with open("example.txt", "r") as file:
    content = file.read()  # Чтение содержимого файла
    print(content)
````

````python
with open('input.txt', "r") as input_file:
    content = input_file.read()  # Чтение содержимого файла
    split_content = content.split() # Разделили строку на массив из слов
    reversed_split_content = split_content[::-1] # сделали порядок слов обратным
    joined_reversed_split_content = " ".join(reversed_split_content) # собрали строку заново, добавив пробелы между словами

    with open("output.txt", "w") as output_file:
        output_file.write(joined_reversed_split_content) # Записали в другой файл
````
