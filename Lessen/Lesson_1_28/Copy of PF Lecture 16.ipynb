## Списковые включения

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = [num*num for num in numbers]
squared_numbers
```

```python
numbers = [1, 2, 3, 4, 5]
squared_simple_numbers = []

for num in numbers:
    squared_simple_numbers.append(num * num)

squared_simple_numbers
```

```python
potential_list_of_lists = [3, [5,3], 7]

for element in potential_list_of_lists:
    if type(element) == list:
        for subelement in element:
            print('внутренний список: ', element, 'подэлемент: ', subelement)
    else:
        print(element)
# чтобы открывать списки с бесконечной(неполностью бесконечной) глубиной, нам нужна рекурсия
```

```python
вводная_строка = '(()))'
```

```python
стэк = []

for скобка in вводная_строка:
    if скобка == ')':
        if len(стэк) == 0 or стэк[-1] != '(':
            print('Последовательность neправильная')
            break
        else:
            стэк.pop()
    else:
        стэк.append(скобка)


if len(стэк) == 0:
    print('Последовательность правильная')
else:
    print('Последовательность neправильная')
```
