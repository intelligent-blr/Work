- Напишите программу, которая сортирует список строк по длине строк с использованием метода sort().
- Если строки имеют одинаковую длину, то они должны быть упорядочены в лексикографическом порядке.
- Программа должна выводить промежуточные этапы сортировки и итоговый отсортированный список.

Пример вывода:

- Введите список строк, разделенных пробелами: apple pear banana cat dog
- Исходный список: ['apple', 'pear', 'banana', 'cat', 'dog']
- Отсортированный список: ['cat', 'dog', 'pear', 'apple', 'banana']


```python
список = ['ape', 'apple', 'pear', 'banana', 'cat', 'dog']
список
```

```python
список.sort(key=lambda строка: (len(строка), строка)) # первоочередно сортируем по нарастанию используя
# длину строки, если длина строки одинаковая с какой то другой, то идет сравнение лексикографическое
```
