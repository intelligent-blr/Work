```python
входные_данные = [1, 2, 8, -3, 5, 7, 11, -2]
```

```python
множество = set()

for число in входные_данные:
    if -число in множество:
        print('Нашли два числа, которые в сумме дают ноль: ', число, ' ', -число)
        break
    множество.add(число)
    print('Не нашли пока то самое число, просто добавили ', число)

множество
```

```python
дубликатный_список = [1, 2, 8, -3, 5, 7, 11]
```

```python
def содержит_дубликаты(список):
    return len(список) == len(set(список))
```

```python
содержит_дубликаты(дубликатный_список)
```
