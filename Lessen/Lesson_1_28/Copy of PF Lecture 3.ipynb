n школьников делят k яблок поровну, неделящийся остаток остается в корзинке. Сколько яблок достанется каждому школьнику? Сколько яблок останется в корзинке? Программа получает на вход числа n и k и должна вывести искомое количество яблок (два числа).

```python
n = int(input('n: '))
k = int(input('k: '))
print('Каждому школьнику достанется: ', k // n, ' яблокa.')
print('В корзине останется: ', k % n, ' яблок(a).')
```

```python
x = 10  # 1010 в двоичной системе
y = 5  # 0101 в двоичной системе

print(x & y)  # Выводит 0
print(x | y)  # Выводит 15
print(x ^ y)  # Выводит 15
print(x << 1)  # Выводит 20
print(x >> 1)  # Выводит 5
print(x ^ x) # используется в задачке с единственным уникальным элементов в массиве
```

```python
arr = [5, 10, 5, 10, 15, 240, 120, 190, 57, 240, 120, 190, 57]
res = 0

for element in arr:
  res ^= element

print(res)
```
