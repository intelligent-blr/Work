```python
a = -6
b = -6
id(a) == id(b)
# id кэшируются от -5 до 256
```

```python
a = 5
b = 5
c = 5
d = 5
e = 5
f = 5
id(a) == id(b) == id(c) == id(d) == id(e) == id(f)
```

```python
a = 4
b = 4
c = 5
id(a) == id(b)  # True
id(a) == id(c)  # False
a += 1
id(a) == id(b)  # False
id(a) == id(c)  # True
```

```python
первое_число = int(input('Введите первое число: '))
второе_число = int(input('Введите второе число: '))
третье_число = int(input('Введите третье число: '))
```

```python
if первое_число == второе_число == третье_число:
  print('Это обман, они не уникальны')
else:
  print('Все три числа уникальные и не равны друг другу.')
```
