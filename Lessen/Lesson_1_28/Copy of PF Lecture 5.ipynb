## Вложенные операторы

```python
x = -1
if x > 0:
  # сюда мы входим
  if x < 10:
    # и сюда мы тоже входим
    print("x находится в диапазоне от 0 до 10")
  else:
    # сюда мы войдем, если мы больше нуля и больше 10
    print("x больше или равно 10")
else:
  print("x меньше или равно 0")
```

```python
# short circuiting
# A and (B | C and G | K)
# A == false => Всё фолс
# A or ......
# A == True => возвращаем тру, не считая то, что справа
```

```python
x = -1
if x:
    print("x истинно")
else:
    print("x ложно")

# 0 = FALSE
# 1 = TRUE
# 256 = TRUE
# -1 = TRUE
```

```python
strokx = 'a'

if strokx:
  print("x истинно")
else:
  print("x ложно")

# пустая строка = FALSE
# любое значение в строке = ИСТИНА
```

### **Практическое задание**

Написать программу, в которой считывается число дней в месяце, в зависимости от этого выводятся все подходящие названия месяцев. Для 28 или 29 также необходимо указать високосность/не високосность года.

```python
колво_дней = int(input('Введите кол-во дней в месяце: '))

if колво_дней == 28 or колво_дней == 29:
  високосность = (input('Високосный ли год? ') == 'Да')
```

```python
месяца_31 = ['январь', 'март', 'май', 'июль', 'август', 'октябрь', 'декабрь']
месяца_30 = ['апрель', 'июнь', 'сентябрь', 'ноябрь']
```

```python
if колво_дней == 31:
  print(месяца_31)
elif колво_дней == 30:
  print(месяца_30)
elif (колво_дней == 28 and not високосность) or (колво_дней == 29 and високосность):
  print('февраль')
else:
  print('нет ни одного месяца')
```

Напишите программу, которая запрашивает у пользователя год его рождения и выводит сообщение о том, к какому поколению он относится.

* "Бэби-бумеры": родившиеся до 1965 года включительно.

* "Поколение Z": родившиеся после 1996 года.

Пример вывода:

* Введите год вашего рождения: 1990
* Вы относитесь к поколению Y (миллениалы).
