```python
# Для трехзначных чисел оставить только те, сумма цифр в которых равна 15.
# Для четных выводить дополнительно последнюю цифру.
# Для оканчивающихся на ноль печатать строку zero.
# Попробовать решить сначала без break/continue, потом избегая вложенных условий.
```

```python
# 100 - 999
# декомпозируем каждое число, сделав его строкой и взял каждую цифру по отдельности
число = 100

while число < 120:
  сумма = 0
  строка = str(число)

  for цифра in строка:
    сумма += int(цифра)

  if сумма != 15:
    print('не делится на 15: ', число)
  if число % 2 == 0: # проверка на четность
    print('четное число: ', число, ' последняя цифра: ', строка[-1])
  if строка[-1] == '0':
    print('число ', число, ' оканчивается на ноль, поэтому ZERO')
  число += 1
```

```python
# Напишите программу, которая запрашивает у пользователя положительное целое число и вычисляет его факториал.
# Факториал числа N вычисляется как произведение всех целых чисел от 1 до N. Используйте цикл с предусловием while для решения задачи.
# Выведите результат на экран с помощью команды print.
```

```python
# Пример вывода:

# Введите положительное целое число: 5
# Факториал числа 5 равен 120.
число = int(input())
результат = число

while число > 1:
  результат *= число - 1
  число -= 1

print(результат)
```

```python
число = int(input())
результат = 1

for i in range(1, число+1): # сработает от 1 до числа раз, если я ввел 5, код сработает 5 раз
# сначала он перемножит на 1 * 1 * 2 * 3 * 4 * 5
  результат *= i
  print(результат)
```
