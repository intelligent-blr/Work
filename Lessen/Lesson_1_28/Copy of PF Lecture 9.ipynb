## **Знакомство со строками**

**Строка** — это упорядоченная последовательность символов, которые используются для хранения и представления текстовой информации.

Примеры строк:



```
'Hello'
'Goodbye'
'G'
' '
''
'Hello, \'World\' '
```

Строки в Python могут быть записаны с использованием кавычек:
*   одинарных (')
*   двойных (")
*   тройных кавычек (""")





````python
# Создание строки с кавычками в тексте
# ' " "   '
print('Dragon\'s mother said "No"')
# => Dragon's mother said "No"
a = 7
a # значение в последней строке автоматически выводится в консольке
````

## **Арифметические операции над строками**

Строки поддерживают операции сложения и умножения.
* Операция сложения позволяет объединять строки
* Операция умножения позволяет повторять строку определенное количество раз

````python
greeting = "Hello, "
name = "John "
message = greeting + name  # Конкатенация строк
multiplied_string = name * 3  # Повторение строки
message, multiplied_string
````

## **Взятие среза от строки**

С помощью срезов можно выбрать определенный диапазон символов.

Варианты срезов включают использование одного, двух или трех аргументов:

* Один аргумент указывает индекс символа, с которого начинается срез.
* Два аргумента указывают начальный и конечный индексы среза.
* Три аргумента позволяют указать шаг, с которым будут выбраны символы.


````python
text = "Hello, world!"
text[3:8]  # Взятие среза с 1-го по 5-й символы
````

````python
text[7:-1]  # Взятие среза с 8-го символа до конца строки
````

````python
text[3:3]  # Взятие среза с шагом 2
````

````python
text = "Hello"
copy = text[:]  # Создание копии строки
reversed_text = text[::-1]  # Разворот
copy, reversed_text
````

## **Магия** - это все команды, которым предшествует знак %
* Линейная магия, то есть магия, которая затрагивает только одну линию кода, предшествуется одним знаком %
* Ячеячная магия, предшествуется двумя знаками % и затрагивает всю ячейку

````python
rep_times = int(input())
````

````python
rep_times_bro = int(input())
````

````python
# for-loop - это конструкция, которая позволяет исполнить тот или иной код определенное кол-во раз
# range - диапазон значений, по умолчанию, началом диапазона является 0, а мы передаем только конец диапазона
# range(rep_times) означает 0..<rep_times диапазон значений, в котором у нас меняется индекс
# диапазон от 0..<5 будет принимать значения: 0, 1, 2, 3, 4
for i in range(rep_times):
  print('наступил внутренний фор луп')
  for i in range(rep_times_bro): # этот фор луп наступит два раза
    print("кусок кода внутри двух фор лупов")
````
