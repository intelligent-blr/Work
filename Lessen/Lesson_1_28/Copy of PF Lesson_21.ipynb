Напишите функцию merge_dicts, которая принимает произвольное количество словарей в качестве аргументов и возвращает новый словарь, объединяющий все входные словари. Если ключи повторяются, значения должны быть объединены в список.

Пример ввода:
```python
{'a': 1, 'b': 2}
{'b': 3, 'c': 4}
{'c': 5, 'd': 6}
```
Пример вывода:
```python
{'a': [1], 'b': [2, 3], 'c': [4, 5], 'd': [6]}
```

````python
def merge_dicts(*args):
    print(args)
    k = {}
    for element in args:
        for key, value in element.items():
            if key in k:
                k[key].append(value)
            else:
                k[key] = [value]
    return k

merge_dicts({'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6})
````

### Практическое задание 1

Напишите функцию group_by_first_letter(words), которая принимает список строк words и возвращает словарь, где:

Ключ — первая буква строки (в нижнем регистре).
Значение — список строк из исходного списка, начинающихся с этой буквы.


````python
def group_by_first_letter(words: list) -> dict:
    result = {}
    for word in words:
        key = word[0].lower()  # берем первую букву в нижнем регистре
        result.setdefault(key, []).append(word)
    return result

words = ["Apple", "apricot", "banana", "Blueberry", "cherry", "Avocado"]
grouped = group_by_first_letter(words)
print(grouped)
````

## Counter

````python
from collections import Counter

# Создаем Counter из списка
lst = ['a', 'b', 'c', 'a', 'b', 'a']
counter = Counter(lst)
print(counter)
````

````python
# Можно также создать Counter из строки
s = 'abracadabra'
counter = Counter(s)
print(counter)
````

````python
# Получение количества вхождений элемента
print(counter['a'])
````

````python
# Получение списка всех уникальных элементов
print(list(counter.keys()))
````

````python
# Получение списка всех значений (количество вхождений)
print(list(counter.values()))
````

````python
# Метод elements() возвращает итератор, повторяющий каждый элемент в Counter столько раз, сколько он появляется в Counter.
print(list(counter.elements()))
````

````python
# Метод most_common() возвращает наиболее часто встречающиеся элементы в порядке убывания количества
print(counter.most_common(2))
````

````python
# Обновление Counter другими счетчиками
counter2 = Counter('aaaaabbbcccd')
counter.update(counter2)
print(counter)
````

# Namedtuple

Именованные кортежи (namedtuple) — это подкласс кортежей в Python. У них те же функции, что и у обычных, но их значения можно получать как с помощью имени (через точку, например, .name), так и с помощью индекса (например [0]). В этом материале на примерах разберем синтаксис и функции этих кортежей.

https://pythonru.com/uroki/namedtuple-imennye-kortezhi-dlja-nachinajushhih

````python
marks = (98, 80, 95)
print(marks)
print(marks[0])
print(marks[2])
````

Namedtuple — это замена обычных кортежей. Они выполняют те же функции, но улучшают читаемость кода.

### Импорт именованных кортежей в Python

По сравнению с другими структурами данных в Python (список, словарь, кортеж и множество) именованный кортеж — не часть стандартной библиотеки. Он доступен в модуле collections. Последний включает и другие альтернативы встроенных структур в Python.

````python
from collections import namedtuple
````

````python
Marks = namedtuple('Marks', 'Physics Chemistry Math English')
marks = Marks(90, 85, 95, 100)
print(marks)
````

В первую очередь объявляется сигнатура кортежа,
* где первый параметр принимает имя типа,
* а второй — названия полей.

Имя типа — это строка с типом, который нужно сделать именованным кортежем.

В качестве второго параметра можно использовать итерируемую структуру данных: список, словарь, кортеж или множество.
Если же это строка, то разные имена нужно разделить с помощью отдельного символа.

### Создание namedtuple с помощью списка


````python
lst = ['Physics', 'Chemistry', 'Math', 'English']
Marks = namedtuple('Marks', lst)
marks = Marks(90, 85, 95, 100)
print(marks)
````

### Создание namedtuple с помощью словаря

Стартовое значение словаря — 0, поскольку значение поля игнорируется кортежем.



````python
dct = {'Physics': 0, 'Chemistry': 0, 'Math': 0, 'English': 0}
Marks = namedtuple('Marks', dct)
marks = Marks(90, 85, 95, 100)
print(marks)
````

### Создание namedtuple с помощью кортежа

````python
tupl = ('Physics', 'Chemistry', 'Math', 'English')
Marks = namedtuple('Marks', tupl)
marks = Marks(90, 85, 95, 100)
print(marks)
````

### Создание namedtuple с помощью множества

````python
subject_set = {'Physics', 'Chemistry', 'Math', 'English'}
Marks = namedtuple('Marks', subject_set)
marks = Marks(90, 85, 95, 100)
print(marks)
````

### Доступ к именам полей namedtuple в python

Дальше рассмотрим, как получать доступ к полям в именованном кортеже. Для этого есть несколько способов:

* Через смещение.
* Через точку.
* getattr().

Доступ к полям можно получить тем же способом, что и в кортежах — с помощью индекса.



````python
print(marks[0])
print(marks[3])
````

Также можно использовать названия полей по аналогии с атрибутами экземпляра класса.



````python
print(marks.Physics)
print(marks.English)
````

Наконец, поля можно получить и с помощью функции getattr()

````python
getattr(marks, "Physics")
getattr(marks, "English")
````

### Практическое задание 1

В магазине хранится список товаров, где каждый товар характеризуется:

name — название товара,

price — цена за единицу,

quantity — количество товара на складе.

Требуется:

Создать тип Product с помощью namedtuple с полями name, price и quantity.
Заполнить список несколькими товарами.

Написать функцию, которая вычисляет для каждого товара его общую стоимость (то есть произведение price на quantity) и находит товар с максимальной общей стоимостью. Вывести название товара с максимальной общей стоимостью и саму эту стоимость.

````python
# 1. Создаём namedtuple для товара
Product = namedtuple('Product', ['name', 'price', 'quantity'])

# 2. Заполняем список товаров
products = [
    Product('Хлеб', 30, 50),
    Product('Молоко', 50, 20),
    Product('Яйца', 10, 100),
    Product('Масло', 150, 10)
]

# 3. Функция для нахождения товара с максимальной общей стоимостью
def find_max_total_product(products):
    max_product = max(products, key=lambda p: p.price * p.quantity)
    max_total = max_product.price * max_product.quantity
    return max_product, max_total


max_product, max_total = find_max_total_product(products)
print(f"Товар с максимальной общей стоимостью: {max_product.name}")
print(f"Общая стоимость товара: {max_total}")
````
