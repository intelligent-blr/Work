## Работа с файлами `with open`

````python
lines = [
    "Это первая строка.",
    "Это вторая строка.",
    "Это третья строка."
]

# Открываем файл для записи
with open("example.txt", "w") as file:
    # Записываем каждую строку из списка в файл
    for line in lines:
        file.write(line + "\n")
````

````python
def reverse_words(input_file, output_file):
    try:
        with open(input_file, "r") as f:
            lines = f.readlines()

        reversed_lines = []

        for line in lines:  # Обрабатываем каждую строку
            reversed_lines.append(line.split()[::-1])

        with open(output_file, "w") as f:
            for line in reversed_lines:
                for word in line:
                    f.write(word + '\n')
    except FileNotFoundError:
        print("Один из файлов не найден.")

reverse_words("input.txt", "output.txt")
````

# Задание 1

Напишите программу, которая будет считывать содержимое текстового файла с именем "data.txt", содержащего числа, разделенные пробелами. Программа должна вычислить сумму всех чисел в файле и записать эту сумму в другой файл с именем "result.txt". Обработайте возможные исключения, такие как FileNotFoundError и ValueError.

Дополнительное усложнение: попробуйте обработать случай, если содержимое файла пусто, и выведите соответствующее сообщение в файл "result.txt".

````python
def process_numbers():
    try:
        with open("data.txt", "r") as infile:
            content = infile.read().strip()
            if not content:
                raise ValueError("Файл пустой")

            # Разбиваем содержимое токены (числа)
            tokens = content.split()

            # Преобразуем токены в числа и вычисляем сумму
            total = 0
            for token in tokens:
                try:
                    number = float(token)
                    total += number
                except ValueError:
                    raise ValueError(f"Неверный формат числа: '{token}'")

            result_text = f"Сумма чисел: {total}"

    except FileNotFoundError:
        result_text = "Файл data.txt не найден."
    except ValueError as e:
        result_text = str(e)

    with open("result.txt", "w") as outfile:
        outfile.write(result_text)
    print("Результат записан в result.txt")


if __name__ == "__main__":
    process_numbers()
````

## Задание 2

Напишите программу, которая открывает файл, считывает его содержимое и выполняет операции над числами в файле.

Обработайте возможные исключения при открытии файла (FileNotFoundError) и при выполнении операций над числами (ValueError, ZeroDivisionError).
Используйте конструкцию try-except-finally для обработки исключений и закрытия файла в блоке finally.


````python
def read_data(file_name):
    f = None
    try:
        f = open(file_name, "r")
        data = f.read().strip()
        numbers = data.split()

        if len(numbers) < 2:
            raise ValueError("В файле должно быть как минимум два числа.")

        num1 = float(numbers[0])
        num2 = float(numbers[1])

        if num1 <= 0 or num2 <= 0:
            raise ValueError("Число должно быть положительным.")

        print("Сумма:", num1 + num2)
        print("Разность:", num1 - num2)
        print("Произведение:", num1 * num2)
        print("Деление:", num1 / num2)

    except FileNotFoundError:
        print("Файл не найден!")
    except ValueError as e:
        print("Ошибка значения:", e)
    except ZeroDivisionError:
        print("Ошибка: деление на ноль!")
    finally:
        if f:
            f.close()
            print("Файл закрыт.")


if __name__ == '__main__':
    file_name = input("Введите имя файла: ")
    read_data(file_name)
````
