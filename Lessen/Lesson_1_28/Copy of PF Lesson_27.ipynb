```python
def get_chars(text: str) -> set[str]:
    alphabet = [chr(i) for i in range(ord("а"), ord("я") + 1)] + [chr(i) for i in range(ord("a"), ord("z") + 1)]
    chars = []
    for chr_ in text.lower():
        if chr_ not in alphabet and chr_ != ' ':
            chars.append(chr_)
    return set(chars)
```

```python
def filter_words(file_path: str) -> list[str]:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    # print(get_chars(text))
    chars = get_chars(text)
    words = text.split(" ")
    for chr in chars:
        words = list(map(lambda word: word.replace(chr, ''), words))
    return list(filter(lambda word: len(word) >= 9, words))

filter_words("anna-karenina.txt")
```

### Практическое задание 2
Напишите программу, которая принимает список чисел от пользователя и использует функцию map, чтобы преобразовать каждый элемент списка в его квадрат.
Затем программа должна использовать функцию filter, чтобы отфильтровать только те
элементы, которые являются четными числами.
В результате программа должна вывести новый список, содержащий квадраты только четных
чисел.

```python
numbers = [1, 2, 3, 4, 8, 7, 1, 0]

def filter_square_data(numbers: list[int]) -> list[int]:
    # numbers_square = [number ** 2 for number in numbers]
    numbers_square = list(map(lambda number: number ** 2, numbers))
    return numbers_square

filter_square_data(numbers)
```

```python
numbers = [1, 2, 3, 4, 8, 7, 1, 11]

def preproc_number(number: int) -> int:
    if number == 1:
        return 2
    elif number > 10:
        return number ** 2
    else:
        return number - 1

def filter_square_data(numbers: list[int]) -> list[int]:
    # numbers_square = [number ** 2 for number in numbers]
    numbers_square = list(map(preproc_number, numbers))
    return numbers_square

filter_square_data(numbers)
```

```python
numbers = [1, 2, 3, 4, 8, 7, 1, 0]

def filter_square_data(numbers: list[int]) -> list[int]:
    # numbers_square = [number ** 2 for number in numbers]
    numbers_square = list(map(lambda number: number ** 2, numbers))
    result = list(filter(lambda number: number % 2 == 0, numbers_square))
    return result

filter_square_data(numbers)
```

```python
numbers = [1, 2, 3, 4, 8, 7, 1, 0]

def filter_square_data(numbers: list[int]) -> list[int]:
    # numbers_square = [number ** 2 for number in numbers]
    result = list(filter(lambda number: number % 2 == 0, map(lambda number: number ** 2, numbers)))
    return result

filter_square_data(numbers)
```
