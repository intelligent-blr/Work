# Методы __str__ и __repr__


````python
class Car:
    status_engine = False  # Атрибут класса (общий для всех машин)

    def __init__(self, color, brand='Porsche'):
        self.brand = brand
        self.color = color
        self.status_engine = False  # Каждый экземпляр имеет свой статус двигателя

    def engine_on(self):
        print("Двигатель заведён")
        self.status_engine = True

    def engine_off(self):
        print("Двигатель заглушен")
        self.status_engine = False

    def __str__(self):
        """Возвращает строковое представление для пользователя"""
        engine_state = "включён" if self.status_engine else "выключен"
        return f"Автомобиль {self.brand}, цвет {self.color}, двигатель {engine_state}."

    def __repr__(self):
        """Возвращает строку для отладки (разработчикам)"""
        return f"Car(brand='{self.brand}', color='{self.color}', status_engine={self.status_engine})"


car1 = Car("Красный")
car2 = Car("Синий", "BMW")

print(car1)
print(car2)

print(repr(car1))

car1.engine_on()
print(car1)
````

### Практическое задание 1: "Робот-помощник" 🤖

Создать класс Robot с полями имя, год выпуска, уровень заряда (в процентах) и список задач.

Добавить методы:

* get_age(): вычисляет возраст робота.
* charge(): заряжает робота до 100%.
* add_task(task): добавляет задачу в список задач (если уровень заряда больше 10%).
* complete_task(): выполняет первую задачу из списка (уменьшая заряд на 10%).
* __str__() и __repr__() для удобного вывода.
Добавить конструкторы для создания робота с разными начальными параметрами.

````python
from datetime import date

class Robot:
    def __init__(self, name, year, battery=100, tasks=None):
        self.name = name
        self.year = year
        self.battery = battery
        self.tasks = tasks if tasks else []

    def get_age(self):
        return date.today().year - self.year

    def charge(self):
        self.battery = 100
        return f"{self.name} полностью заряжен!"

    def add_task(self, task):
        if self.battery > 10:
            self.tasks.append(task)
            return f"Задача '{task}' добавлена!"
        return f"Недостаточно заряда для добавления задачи!"

    def complete_task(self):
        if self.tasks and self.battery > 10:
            task = self.tasks.pop(0)
            self.battery -= 10
            return f"{self.name} выполнил задачу: {task}. Остаток заряда: {self.battery}%"
        return f"Нет задач или недостаточно заряда!"

    def __str__(self):
        return (f"🤖 Робот {self.name}, возраст: {self.get_age()} лет, "
                f"заряд: {self.battery}%, задач: {len(self.tasks)}")

    def __repr__(self):
        return f"Robot(name={self.name!r}, year={self.year}, battery={self.battery}, tasks={self.tasks})"

robot1 = Robot("R2-D2", 2010, 50)
robot2 = Robot("Bender", 2022)

print(robot1)
print(robot1.add_task("Принести кофе"))
print(robot1.complete_task())
print(robot1.charge())
print(robot1)

print(repr(robot2))
````

### Практическое задание 2: Класс `Person` 🧑

#### **Описание задачи**
Создайте класс `Person`, который описывает человека с основными характеристиками:
- **Имя** (`name`)
- **Год рождения** (`birth_year`)
- **Рост** (`height`, в метрах, необязательный параметр)
- **Вес** (`weight`, в килограммах, необязательный параметр)

Добавьте методы:
1. `get_age()`: вычисляет возраст человека на основе текущего года.
2. `get_bmi()`: рассчитывает индекс массы тела (ИМТ) по формуле **вес / (рост²)**. Если рост и вес не указаны, возвращает `None`.

Реализуйте:
- Разные конструкторы: объект можно создавать с разным количеством параметров.
- Переопределите методы `__str__` и `__repr__`:
  - `__str__()` — красивый вывод информации о человеке.
  - `__repr__()` — строковое представление объекта для отладки.

#### **Пример работы кода**
```python
p1 = Person("Иван", 1990, 1.75, 70)
p2 = Person("Мария", 1985)

print(p1)   # Имя: Иван, Возраст: 34 лет, ИМТ: 22.86
print(repr(p1))  # Person(name='Иван', birth_year=1990, height=1.75, weight=70)
print(p2)   # Имя: Мария, Возраст: 39 лет, ИМТ: не указан
print(repr(p2))  # Person(name='Мария', birth_year=1985, height=None, weight=None)
```

### Практическое задание 3: Класс "Космический корабль" 🚀  

Создай класс **`Spaceship`**, который моделирует космический корабль. У корабля должны быть:  
- Имя корабля  
- Текущий запас топлива (в литрах)  
- Максимальный запас топлива  
- Скорость (изначально 0)  

### **Методы:**
1. **`refuel(amount)`** – заправляет корабль, но не больше максимального запаса топлива.  
2. **`accelerate(speed_increase, fuel_cost)`** – увеличивает скорость корабля, но требует топлива. Если топлива не хватает – выдать предупреждение.  
3. **`brake(amount)`** – снижает скорость корабля, но не ниже 0.  
4. **`__str__` и `__repr__`** – возвращают строковое представление корабля.  

### **Дополнительно:**  
- Добавь метод **`travel(time)`**, который рассчитывает пройденное расстояние (скорость * время).  
- Введи ограничение максимальной скорости.  

**Пример использования:**  
```python
falcon = Spaceship("Falcon-9", max_fuel=5000)

falcon.refuel(3000)
falcon.accelerate(1000, 500)
print(falcon)  # Корабль Falcon-9: скорость 1000, топливо 2500

falcon.brake(500)
print(falcon)  # Корабль Falcon-9: скорость 500, топливо 2500
```
