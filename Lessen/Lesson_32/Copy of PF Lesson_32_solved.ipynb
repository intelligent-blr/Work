### Магический метод add

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency

    def add_money(self, other):
        """
        Складывает два объекта Money, если их валюты совпадают.
        Возвращает новый объект Money с суммарной суммой.
        """
        if self.currency == other.currency:
            return Money(self.amount + other.amount, self.currency)
        else:
            raise ValueError("Валюты не совпадают")

    def __str__(self):
        """
        Возвращает строковое представление объекта Money.
        """
        return f"{self.amount} dfsfsdf {self.currency}"

    def __repr__(self) -> str:
        return "dsfaslf;lasfj;lsajf;jsa;dfljla;ksjf;sjaf"

m1 = Money(100, "eur")
m2 = Money(500, "eur")
```

### Реализуем остальные методы

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency

    def __add__(self, other):
        """
        Складывает два объекта Money, если валюты совпадают.
        """
        if isinstance(other, Money) and self.currency == other.currency:
            return Money(self.amount + other.amount, self.currency)
        raise ValueError("Нельзя складывать разные валюты")

    def __sub__(self, other):
        """
        Вычитает один объект Money из другого, если валюты совпадают.
        """
        if isinstance(other, Money) and self.currency == other.currency:
            return Money(self.amount - other.amount, self.currency)
        raise ValueError("Нельзя вычитать разные валюты")

    def __mul__(self, factor):
        """
        Умножает объект Money на число.
        """
        if isinstance(factor, (int, float)):
            return Money(self.amount * factor, self.currency)
        raise TypeError("Умножать можно только на число")

    def __truediv__(self, factor):
        """
        Делит объект Money на число.
        """
        if isinstance(factor, (int, float)):
            if factor == 0:
                raise ZeroDivisionError("Деление на ноль невозможно")
            return Money(self.amount / factor, self.currency)
        raise TypeError("Делить можно только на число")

    def __floordiv__(self, factor):
        """
        Целочисленное деление объекта Money на число.
        """
        if isinstance(factor, (int, float)):
            if factor == 0:
                raise ZeroDivisionError("Деление на ноль невозможно")
            return Money(self.amount // factor, self.currency)
        raise TypeError("Делить можно только на число")

    def __mod__(self, factor):
        """
        Остаток от деления суммы денег на число.
        """
        if isinstance(factor, (int, float)):
            return Money(self.amount % factor, self.currency)
        raise TypeError("Остаток от деления можно вычислять только с числом")

    def __pow__(self, exponent):
        """
        Возведение суммы в степень.
        """
        if isinstance(exponent, (int, float)):
            return Money(self.amount ** exponent, self.currency)
        raise TypeError("Степень должна быть числом")

    def __str__(self):
        """
        Возвращает строковое представление объекта Money.
        """
        return f"{self.amount} {self.currency}"


m1 = Money(100, "eur")
m2 = Money(50, "eur")

print(m1 + m2)  # 150 eur
print(m1 - m2)  # 50 eur
print(m1 * 2)   # 200 eur
print(m1 / 2)   # 50.0 eur
print(m1 // 3)  # 33 eur
print(m1 % 30)  # 10 eur
print(m1 ** 2)  # 10000 eur

m3 = Money(100, "usd")
print(m1 + m3)
```

### Реализуем операторы сравнения для класса

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency

    def __eq__(self, other):
        """Оператор == (равенство)"""
        if isinstance(other, Money) and self.currency == other.currency:
            return self.amount == other.amount
        return False

    def __ne__(self, other):
        """Оператор != (неравенство)"""
        return not self.__eq__(other)  # Используем __eq__

    def __lt__(self, other):
        """Оператор < (меньше)"""
        if isinstance(other, Money) and self.currency == other.currency:
            return self.amount < other.amount
        raise ValueError("Нельзя сравнивать разные валюты")

    def __le__(self, other):
        """Оператор <= (меньше или равно)"""
        return self.__lt__(other) or self.__eq__(other)  # Используем < и ==

    def __gt__(self, other):
        """Оператор > (больше)"""
        return not self.__le__(other)  # Используем <=

    def __ge__(self, other):
        """Оператор >= (больше или равно)"""
        return not self.__lt__(other)  # Используем <

    def __str__(self):
        return f"{self.amount} {self.currency}"


m1 = Money(100, "usd")
m2 = Money(200, "usd")
m3 = Money(100, "usd")
m4 = Money(50, "eur")

print(m1 == m3)  # True (оба 100 usd)
print(m1 != m2)  # True (100 usd != 200 usd)
print(m1 < m2)   # True (100 usd < 200 usd)
print(m2 > m1)   # True (200 usd > 100 usd)
print(m1 <= m3)  # True (100 usd <= 100 usd)
print(m2 >= m1)  # True (200 usd >= 100 usd)

# Попытка сравнить разные валюты
try:
    print(m1 > m4)
except ValueError as e:
    print(e)
```

### Вспомним сэттеры и геттеры

```python
class Money:
    def __init__(self, amount, currency):
        self._amount = amount
        self._currency = currency

    def get_amount(self):
        """
        Геттер: возвращает текущую сумму и валюту в виде словаря.
        """
        return {self._currency: self._amount}

    def set_amount(self, value):
        """
        Сеттер: принимает словарь формата {"usd": 100}.
        Если валюта совпадает, обновляет сумму; иначе выводит предупреждение.
        """
        if not isinstance(value, dict) or len(value) != 1:
            raise ValueError("Ожидается словарь формата {'usd': 100}")

        (currency, amount) = list(value.items())[0]

        if currency == self._currency:
            self._amount = amount
            print("Успешно!")
        else:
            print("Валюта не совпадает")

    def __str__(self):
        """
        Преобразование объекта в строку.
        """
        return f"{self._amount} {self._currency}"


m = Money(1000, "usd")

m.set_amount({"usd": 100})
m.set_amount({"eur": 200})

print(m.get_amount())

print(m)
```

### setitem and getitem

```python
class Money:
    def __init__(self, amount, currency):
        self._amount = amount
        self._currency = currency

    def __getitem__(self, currency):
        """
        Геттер: возвращает сумму, если валюта совпадает, иначе -1.
        """
        if currency == self._currency:
            return self._amount
        return -1

    def __setitem__(self, currency, amount):
        """
        Сеттер: изменяет сумму, если валюта совпадает, иначе выводит предупреждение.
        """
        if currency == self._currency:
            self._amount = amount
            print("Успешно!")
        else:
            print("Валюта не совпадает")

    def __str__(self):
        """
        Преобразование объекта в строку.
        """
        return f"{self._amount} {self._currency}"


m = Money(1000, "usd")

m["usd"] = 100
m["eur"] = 200

print(m["usd"])
print(m["eur"])

print(m)
```

### Создаём свой итератор

```python
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.start >= self.end:
            raise StopIteration  # Вызываем исключение StopIteration
        current = self.start
        self.start += 1
        return current

numbers = MyRange(1, 5)
iterator = iter(numbers)  # Получаем итератор
```

### Пример 1

```python
import time
import random

class ChatStream:
    def __init__(self, users):
        self.users = users  # Список пользователей
        self.messages = []  # Хранилище сообщений
        self.last_index = 0  # Индекс последнего прочитанного сообщения

    def add_message(self, user, message):
        """Добавляет новое сообщение в чат."""
        self.messages.append(f"{user}: {message}")

    def __iter__(self):
        return self  # Итератором является сам объект

    def __next__(self):
        """Ждёт появления новых сообщений и возвращает их по мере поступления."""
        while self.last_index >= len(self.messages):  # Ждём новое сообщение
            time.sleep(1)  # Имитация ожидания нового сообщения

        message = self.messages[self.last_index]  # Берём следующее сообщение
        self.last_index += 1  # Увеличиваем индекс
        return message

chat = ChatStream(["Alice", "Bob", "Charlie"])
print("Ожидание сообщений в чате...")

for i in range(10):  # Симуляция 10 сообщений
    sender = random.choice(chat.users)
    message = f"Сообщение {random.randint(1, 100)}"

    chat.add_message(sender, message)
    print("Новое сообщение:", next(chat))

    time.sleep(random.uniform(0.5, 2))  # Ждём перед отправкой следующего
```
